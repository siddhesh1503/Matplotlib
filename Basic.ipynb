```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri',]  #x-axis
y = [10, 15, 7, 20, 12]

plt.plot(x, y)

plt.title('Bakery Sales of This Week')
plt.xlabel('Day of the week')
plt.ylabel('Sales per Day')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Taking input from the user
x_values = input("Enter x values separated by space: ")
y_values = input("Enter y values separated by space: ")

# Convert the input strings into lists of numbers
x = list(map(float, x_values.split()))
y = list(map(float, y_values.split()))

# Plotting the graph
plt.plot(x, y, marker='o')

# Adding titles and labels
plt.title("User Input Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```
